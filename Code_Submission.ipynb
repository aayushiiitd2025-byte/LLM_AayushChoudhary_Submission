```python


!pip install -q transformers accelerate bitsandbytes pandas matplotlib seaborn tqdm


# =========================================================
# CELL 2 — IMPORTS
# =========================================================

import json
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
import torch

from tqdm import tqdm
from collections import Counter

from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    pipeline
)


# =========================================================
# CELL 3 — LOAD DATASET
# =========================================================

url = "https://raw.githubusercontent.com/czyssrs/FinQA/main/dataset/test.json"

data = json.loads(requests.get(url).text)

print("Total examples:", len(data))

sample = data[0]

print("\nQUESTION:")
print(sample['qa']['question'])

print("\nPROGRAM:")
print(sample['qa']['program'])

print("\nANSWER:")
print(sample['qa']['exe_ans'])

print("\nGOLD EVIDENCE:")
print(sample['qa']['gold_inds'])


# =========================================================
# CELL 4 — ANALYSIS FUNCTIONS
# =========================================================

def get_reasoning_depth(program):

    ops = re.findall(
        r"(add|subtract|multiply|divide)\(",
        program
    )

    return len(ops)


def evidence_type(gold_inds):

    has_table = False
    has_text = False

    for g in gold_inds:

        if "table" in g:
            has_table = True

        if "text" in g:
            has_text = True

    if has_table and has_text:
        return "mixed"

    elif has_table:
        return "table_only"

    elif has_text:
        return "text_only"

    return "unknown"


# =========================================================
# CELL 5 — REASONING DEPTH DISTRIBUTION
# =========================================================

reasoning_depths = []

for entry in data:

    prog = entry['qa']['program']

    reasoning_depths.append(
        get_reasoning_depth(prog)
    )

counter = Counter(reasoning_depths)

print(counter)


# =========================================================
# CELL 6 — PLOT REASONING DEPTH
# =========================================================

plt.figure(figsize=(6,4))

plt.bar(counter.keys(), counter.values())

plt.xlabel("Reasoning Depth")
plt.ylabel("Count")
plt.title("Reasoning Chain Distribution")

plt.show()


# =========================================================
# CELL 7 — EVIDENCE DISTRIBUTION
# =========================================================

evidence_counts = []

for entry in data:

    gold_inds = entry['qa']['gold_inds']

    evidence_counts.append(
        evidence_type(gold_inds)
    )

evidence_counter = Counter(evidence_counts)

print(evidence_counter)


# =========================================================
# CELL 8 — PLOT EVIDENCE DISTRIBUTION
# =========================================================

plt.figure(figsize=(6,6))

plt.pie(
    evidence_counter.values(),
    labels=evidence_counter.keys(),
    autopct='%1.1f%%'
)

plt.title("Evidence Source Distribution")

plt.show()


# =========================================================
# CELL 9 — PERCENTAGE QUESTIONS
# =========================================================

percentage_examples = []

for entry in data:

    q = entry['qa']['question'].lower()

    if "percentage" in q or "%" in q:
        percentage_examples.append(entry)

print("Percentage Questions:", len(percentage_examples))

for ex in percentage_examples[:10]:

    print("\nQUESTION:")
    print(ex['qa']['question'])

    print("PROGRAM:")
    print(ex['qa']['program'])


# =========================================================
# CELL 10 — LOAD QWEN MODEL
# =========================================================

model_name = "Qwen/Qwen2.5-0.5B-Instruct"

tokenizer = AutoTokenizer.from_pretrained(model_name)

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    device_map="auto",
    torch_dtype=torch.float16
)


# =========================================================
# CELL 11 — GENERATION PIPELINE
# =========================================================

pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer
)


# =========================================================
# CELL 12 — TABLE FORMATTER
# =========================================================

def format_table(table):

    rows = []

    for row in table[:10]:

        rows.append(" | ".join(row))

    return "\n".join(rows)


# =========================================================
# CELL 13 — PROMPT BUILDER (FIXED)
# =========================================================

def build_prompt(entry, mode="full"):

    table_str = format_table(entry['table'])

    pre_text = " ".join(entry['pre_text'][-3:])
    post_text = " ".join(entry['post_text'][:3])

    context_text = pre_text + " " + post_text

    question = entry['qa']['question']

    if mode == "table_only":
        context_text = ""

    elif mode == "text_only":
        table_str = ""

    prompt = f"""
You are a financial reasoning system.

You MUST output ONLY in this exact format.

Program: subtract(5829,5735)
Answer: 94

Allowed operations:
add
subtract
multiply
divide

Table:
{table_str}

Context:
{context_text}

Question:
{question}
"""

    return prompt


# =========================================================
# CELL 14 — GENERATE RESPONSE (FIXED)
# =========================================================

def generate_response(prompt):

    outputs = pipe(
        prompt,
        max_new_tokens=64,
        do_sample=False,
        return_full_text=False
    )

    return outputs[0]['generated_text']


# =========================================================
# CELL 15 — PARSERS (FIXED)
# =========================================================

def parse_program(text):

    match = re.search(
        r"Program:\s*(.*?)\s*Answer:",
        text,
        re.DOTALL | re.IGNORECASE
    )

    if match:

        prog = match.group(1).strip()

        prog = prog.replace("\n", " ")

        return prog

    return None


def parse_answer(text):

    match = re.search(
        r"Answer:\s*([-+]?\d*\.?\d+)",
        text,
        re.IGNORECASE
    )

    if match:
        return float(match.group(1))

    return None


# =========================================================
# CELL 16 — PROGRAM EXECUTOR
# =========================================================

def resolve_value(token, memory):

    token = token.strip()

    if token.startswith("#"):
        return memory[token]

    token = token.replace(",", "")

    return float(token)


def execute_program(program):

    memory = {}

    steps = re.findall(
        r"(add|subtract|multiply|divide)\([^\)]*\)",
        program
    )

    final_value = None

    for idx, step in enumerate(steps):

        op_match = re.match(
            r"(\w+)\((.*)\)",
            step
        )

        if not op_match:
            return None

        op = op_match.group(1)

        args = op_match.group(2).split(",")

        if len(args) != 2:
            return None

        a = resolve_value(args[0], memory)
        b = resolve_value(args[1], memory)

        if op == "add":
            val = a + b

        elif op == "subtract":
            val = a - b

        elif op == "multiply":
            val = a * b

        elif op == "divide":

            if b == 0:
                return None

            val = a / b

        else:
            return None

        memory[f"#{idx}"] = val

        final_value = val

    return final_value


# =========================================================
# CELL 17 — ACCURACY FUNCTIONS
# =========================================================

def execution_accuracy(pred, gold):

    try:
        return abs(float(pred) - float(gold)) < 1e-3

    except:
        return False


def program_accuracy(pred_prog, gold_prog):

    if pred_prog is None:
        return False

    return pred_prog.strip() == gold_prog.strip()


# =========================================================
# CELL 18 — ERROR CATEGORIZATION
# =========================================================

def categorize_error(pred_prog, gold_prog, pred_ans, gold_ans):

    if pred_prog is None:
        return "Parsing Error"

    if pred_ans is None:
        return "Parsing Error"

    if execution_accuracy(pred_ans, gold_ans):

        if not program_accuracy(
            pred_prog,
            gold_prog
        ):
            return "Correct Answer Wrong Program"

        return "Correct"

    pred_numbers = re.findall(
        r"-?\d+\.?\d*",
        pred_prog
    )

    gold_numbers = re.findall(
        r"-?\d+\.?\d*",
        gold_prog
    )

    if set(pred_numbers) != set(gold_numbers):
        return "Retrieval Error"

    pred_ops = re.findall(
        r"(add|subtract|multiply|divide)",
        pred_prog
    )

    gold_ops = re.findall(
        r"(add|subtract|multiply|divide)",
        gold_prog
    )

    if set(pred_ops) != set(gold_ops):
        return "Operation Error"

    if pred_ops == gold_ops:
        return "Order Error"

    return "Hallucination"


# =========================================================
# CELL 19 — RUN BASELINE EVALUATION
# =========================================================

results = []

sample_data = data[:100]

for idx, entry in enumerate(tqdm(sample_data)):

    prompt = build_prompt(entry)

    output = generate_response(prompt)

    if idx < 3:
        print("\nMODEL OUTPUT:")
        print(output)
        print("="*80)

    pred_prog = parse_program(output)

    pred_ans = parse_answer(output)

    gold_prog = entry['qa']['program']

    gold_ans = entry['qa']['exe_ans']

    exec_acc = execution_accuracy(
        pred_ans,
        gold_ans
    )

    prog_acc = program_accuracy(
        pred_prog,
        gold_prog
    )

    error_type = categorize_error(
        pred_prog,
        gold_prog,
        pred_ans,
        gold_ans
    )

    results.append({
        'question': entry['qa']['question'],
        'gold_program': gold_prog,
        'pred_program': pred_prog,
        'gold_answer': gold_ans,
        'pred_answer': pred_ans,
        'execution_accuracy': exec_acc,
        'program_accuracy': prog_acc,
        'error_type': error_type
    })


# =========================================================
# CELL 20 — RESULTS DATAFRAME
# =========================================================

results_df = pd.DataFrame(results)

results_df.head()


# =========================================================
# CELL 21 — ACCURACY
# =========================================================

exec_acc = results_df[
    'execution_accuracy'
].mean()

prog_acc = results_df[
    'program_accuracy'
].mean()

print(
    "Execution Accuracy:",
    round(exec_acc * 100, 2),
    "%"
)

print(
    "Program Accuracy:",
    round(prog_acc * 100, 2),
    "%"
)


# =========================================================
# CELL 22 — ERROR DISTRIBUTION
# =========================================================

error_counts = results_df[
    'error_type'
].value_counts()

print(error_counts)


# =========================================================
# CELL 23 — ERROR PIE CHART
# =========================================================

plt.figure(figsize=(8,6))

plt.pie(
    error_counts.values,
    labels=error_counts.index,
    autopct='%1.1f%%'
)

plt.title("Error Taxonomy Distribution")

plt.show()


# =========================================================
# CELL 24 — ACCURACY VS DEPTH
# =========================================================

depth_acc = {}

for idx, entry in enumerate(sample_data):

    depth = get_reasoning_depth(
        entry['qa']['program']
    )

    if depth not in depth_acc:
        depth_acc[depth] = []

    depth_acc[depth].append(
        results[idx]['execution_accuracy']
    )

final_depth_acc = {
    k: np.mean(v)
    for k, v in depth_acc.items()
}

print(final_depth_acc)


# =========================================================
# CELL 25 — DEPTH PLOT
# =========================================================

plt.figure(figsize=(7,5))

plt.bar(
    final_depth_acc.keys(),
    final_depth_acc.values()
)

plt.xlabel("Reasoning Depth")
plt.ylabel("Execution Accuracy")
plt.title("Accuracy vs Reasoning Complexity")

plt.show()


# =========================================================
# CELL 26 — CONTEXT EVALUATION
# =========================================================

def evaluate_mode(mode_name):

    accuracies = []

    for entry in tqdm(sample_data):

        prompt = build_prompt(
            entry,
            mode=mode_name
        )

        output = generate_response(prompt)

        pred_ans = parse_answer(output)

        gold_ans = entry['qa']['exe_ans']

        acc = execution_accuracy(
            pred_ans,
            gold_ans
        )

        accuracies.append(acc)

    return np.mean(accuracies)


# =========================================================
# CELL 27 — RUN CONTEXT EXPERIMENTS
# =========================================================

full_acc = evaluate_mode("full")

table_acc = evaluate_mode("table_only")

text_acc = evaluate_mode("text_only")

print("Full Context Accuracy:", full_acc)

print("Table Only Accuracy:", table_acc)

print("Text Only Accuracy:", text_acc)


# =========================================================
# CELL 28 — CONTEXT PLOT
# =========================================================

modes = [
    'Full',
    'Table Only',
    'Text Only'
]

accs = [
    full_acc,
    table_acc,
    text_acc
]

plt.figure(figsize=(7,5))

plt.bar(modes, accs)

plt.ylabel("Execution Accuracy")

plt.title("Context Sensitivity Analysis")

plt.show()


# =========================================================
# CELL 29 — SYMBOLIC VERIFICATION
# =========================================================

def verified_prediction(output):

    pred_prog = parse_program(output)

    pred_ans = parse_answer(output)

    if pred_prog is None:
        return pred_ans

    try:

        verified_ans = execute_program(
            pred_prog
        )

        if verified_ans is not None:
            return verified_ans

    except:
        pass

    return pred_ans


# =========================================================
# CELL 30 — EVALUATE VERIFICATION
# =========================================================

baseline_correct = 0
verified_correct = 0

for entry in tqdm(sample_data):

    prompt = build_prompt(entry)

    output = generate_response(prompt)

    baseline_ans = parse_answer(output)

    verified_ans = verified_prediction(output)

    gold_ans = entry['qa']['exe_ans']

    if execution_accuracy(
        baseline_ans,
        gold_ans
    ):
        baseline_correct += 1

    if execution_accuracy(
        verified_ans,
        gold_ans
    ):
        verified_correct += 1


# =========================================================
# CELL 31 — VERIFICATION RESULTS
# =========================================================

baseline_acc = baseline_correct / len(sample_data)

verified_acc = verified_correct / len(sample_data)

print("Baseline Accuracy:", baseline_acc)

print("Verified Accuracy:", verified_acc)


# =========================================================
# CELL 32 — VERIFICATION PLOT
# =========================================================

methods = [
    'Baseline',
    'Verified'
]

scores = [
    baseline_acc,
    verified_acc
]

plt.figure(figsize=(6,5))

plt.bar(methods, scores)

plt.ylabel("Execution Accuracy")

plt.title("Impact of Symbolic Verification")

plt.show()


# =========================================================
# CELL 33 — SAVE RESULTS
# =========================================================

results_df.to_csv(
    "finqa_results.csv",
    index=False
)

print("Results saved successfully")


# =========================================================
# CELL 34 — DOWNLOAD RESULTS
# =========================================================

from google.colab import files

files.download("finqa_results.csv")


# =========================================================
# CELL 35 — FINAL SUMMARY
# =========================================================

print("="*60)
print("FINAL SUMMARY")
print("="*60)

print(
    "\nExecution Accuracy:",
    round(exec_acc*100,2),
    "%"
)

print(
    "Program Accuracy:",
    round(prog_acc*100,2),
    "%"
)

print("\nMost Common Error:")

print(error_counts.idxmax())

print("\nKey Findings:")

print(
    "1. Small LLMs struggle with multi-step arithmetic"
)

print(
    "2. Tables are critical for financial reasoning"
)

print(
    "3. Correct answers do not imply correct reasoning"
)

print(
    "4. Symbolic verification improves reliability"
)
```
